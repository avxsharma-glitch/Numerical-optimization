```python
import numpy as np

def vector_operations():
    print("=== VECTOR OPERATIONS ===")

    # Create vectors
    v1 = np.array([2, 4, 6])
    v2 = np.array([1, 3, 5])

    print(f"v1 = {v1}")
    print(f"v2 = {v2}")

    # Various vector operations
    operations = {
        "Addition": v1 + v2,
        "Subtraction": v1 - v2,
        "Element-wise Multiplication": v1 * v2,
        "Dot Product": np.dot(v1, v2),
        "Cross Product (3D)": np.cross(v1, v2) if len(v1) == 3 else "N/A"
    }

    for op_name, result in operations.items():
        print(f"{op_name}: {result}")

def matrix_operations():
    print("\n=== MATRIX OPERATIONS ===")

    # Create matrices
    A = np.array([[1, 2], [3, 4]])
    B = np.array([[5, 6], [7, 8]])

    print("Matrix A:")
    print(A)
    print("Matrix B:")
    print(B)

    # Various matrix operations
    operations = {
        "Addition": A + B,
        "Subtraction": A - B,
        "Element-wise Multiplication": A * B,
        "Matrix Multiplication": A @ B,
        "Transpose of A": A.T,
        "Determinant of A": np.linalg.det(A),
        "Inverse of A": np.linalg.inv(A)
    }

    for op_name, result in operations.items():
        print(f"\n{op_name}:")
        print(result)

# Run the examples
if __name__ == "__main__":
    vector_operations()
    matrix_operations()
```

```python
# Structured Data (CSV)

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Structured CSV
data = pd.DataFrame({
    'Order_ID': [101, 102, 103, 104, 105],
    'Customer': ['Vaibhav', 'Aryan', 'Alok', 'Vishal', 'Khushi'],
    'Product': ['Shoes', 'Watch', 'Bag', 'Laptop', 'Mobile'],
    'Quantity': [2, 1, 3, 1, 2],
    'Price': [2500, 4000, 1500, 55000, 18000],
    'Rating': [4.5, 4.0, 4.8, 4.2, 3.8]
})

print(" Structured Data (E-commerce Sales):")
display(data)

# --- Basic statistical analysis ---
print("\nColumn Names:", data.columns)
print("Shape:", data.shape)

# Descriptive statistics
print("\nSummary Statistics:")
display(data.describe())

# Computing mean, median, standard deviation for numerical columns
print("Average Price:", data['Price'].mean())
print("Median Quantity:", data['Quantity'].median())
print("Standard Deviation of Ratings:", data['Rating'].std())
```

```python
# Semi-Structured Data (JSON)

import json

# Sample JSON - customer details
json_data = '''
[
  {"Customer_ID": 1, "Name": "Vaibhav", "Purchases": ["Shoes", "Bag"], "Location": {"City": "Delhi", "Country": "India"}},
  {"Customer_ID": 2, "Name": "Aryan", "Purchases": ["Laptop", "Mouse"], "Location": {"City": "Mumbai", "Country": "India"}},
  {"Customer_ID": 3, "Name": "Alok", "Purchases": ["Watch", "Mobile"], "Location": {"City": "Pune", "Country": "India"}}
]
'''

# Load JSON
data_json = json.loads(json_data)
df_json = pd.json_normalize(data_json)

print(" Semi-Structured Data (JSON - Customer Info):")
display(df_json)

print("\nFlattened JSON DataFrame:")
print(df_json)
```

```python
# Unstructured Data

# Product Reviews
reviews = [
    "The product is amazing and delivery was on time!",
    "Bad quality and poor customer service.",
    "Excellent product, totally worth the price.",
    "Delivery was late but the packaging was good.",
    "Average product, not too great but okay for the price."
]

# Basic word frequency analysis

print(" Word Frequency Analysis:")
from collections import Counter
words = " ".join(reviews).lower().split()
word_count = Counter(words)
print(" Most Common Words:")
print(word_count.most_common(10))
```

```python
# Sentiment Analysis

from textblob import TextBlob
sentiments = []

for review in reviews:
    blob = TextBlob(review)
    sentiments.append(blob.sentiment.polarity)

sentiment_df = pd.DataFrame({'Review': reviews, 'Sentiment_Score': sentiments})
print("\n Sentiment Analysis Results:")
display(sentiment_df)

# Classify sentiment type
sentiment_df['Sentiment'] = sentiment_df['Sentiment_Score'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
display(sentiment_df)
```
